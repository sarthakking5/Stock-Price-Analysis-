## Data

We need to get data using pandas datareader. We will get stock information for the following banks:
*  Bank of America
* CitiGroup
* Goldman Sachs
* JPMorgan Chase
* Morgan Stanley
* Wells Fargo



```python
start_date = '2006-01-01'
end_date = '2020-1-1'
```

```python
bac = data.DataReader("BAC", 'yahoo', start_date, end_date)
ms=data.DataReader('MS','yahoo',start_date,end_date)
wfc=data.DataReader('WFC','yahoo',start_date,end_date)
c=data.DataReader('C','yahoo',start_date,end_date)
jpm=data.DataReader('JPM','yahoo',start_date,end_date)
gs=data.DataReader('GS','yahoo',start_date,end_date)
axp=data.DataReader('AXP','yahoo',start_date,end_date)
```

```python
#gs
```

**Create a list of the ticker symbols (as strings). Call this list: tickers**

```python
tickers=['BAC','MS','WFC','C','JPM','GS','AXP']
```

```python
bank_stocks=pd.concat([bac,ms,wfc,c,jpm,gs,axp],axis=1,keys=tickers)
```

```python
bank_stocks
```

**Set the column name levels:**

```python
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

**Head of the bank_stocks dataframe.**

```python
bank_stocks.head()
```



**Max Close price for each bank's stock throughout the time period**

```python
for tick in tickers:
     x=bank_stocks[tick]['Close'].max()
     print(tick,x)
```

```python
#multilevelindexing
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns1=pd.DataFrame()
```

** We can use pandas pct_change() method on the Close column to create a column representing this return value. Create a for loop that goes and for each Bank Stock Ticker creates this returns column and set's it as a column in the returns DataFrame.**

```python
for tick in tickers:
    returns1[tick+' Return']=bank_stocks[tick]['Close'].pct_change()
returns1.head()
```
